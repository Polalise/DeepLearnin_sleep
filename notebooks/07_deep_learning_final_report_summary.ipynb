```python
# 07 노트북 1번 셀: 최종 보고 요약용 결과 파일 로드
# 목적:
# - Phase 1A same-date 결과, Phase 2A previous-day 결과, bootstrap 결과, 최종 보고서를 모두 로드한다.
# - 최종 요약/보고서 반영 전에 파일 존재 여부와 shape를 확인한다.

from pathlib import Path
import pandas as pd
import numpy as np

PROJECT_ROOT = Path(r"c:\workSpace\DeepLearnin_sleep")

PHASE_1A_OUTPUT_ROOT = (
    PROJECT_ROOT
    / "data"
    / "processed"
    / "deep_learning_subset_experiments"
    / "phase_1a_outputs"
)

PHASE_2A_OUTPUT_ROOT = (
    PROJECT_ROOT
    / "data"
    / "processed"
    / "deep_learning_previous_day"
    / "experiments"
    / "phase_2a_outputs"
)

REPORT_PATH = (
    PROJECT_ROOT
    / "reports"
    / "deep_learning_subset_ablation_and_previous_day_report.md"
)

paths = {
    "phase_1a_metrics": PHASE_1A_OUTPUT_ROOT / "phase_1a_model_metrics.csv",
    "phase_1a_threshold_metrics": PHASE_1A_OUTPUT_ROOT / "phase_1a_threshold_tuned_metrics.csv",
    "phase_1a_predictions": PHASE_1A_OUTPUT_ROOT / "phase_1a_model_predictions.csv",
    "phase_2a_metrics": PHASE_2A_OUTPUT_ROOT / "phase_2a_previous_day_model_metrics.csv",
    "phase_2a_threshold_metrics": PHASE_2A_OUTPUT_ROOT / "phase_2a_previous_day_threshold_tuned_metrics.csv",
    "phase_2a_predictions": PHASE_2A_OUTPUT_ROOT / "phase_2a_previous_day_model_predictions.csv",
    "same_vs_previous_comparison": PHASE_2A_OUTPUT_ROOT / "same_date_vs_previous_day_threshold_comparison.csv",
    "top_test_candidates": PHASE_2A_OUTPUT_ROOT / "same_date_vs_previous_day_top_test_candidates.csv",
    "bootstrap_summary": PHASE_2A_OUTPUT_ROOT / "selected_candidate_participant_bootstrap_summary.csv",
    "final_report": REPORT_PATH,
}

path_check = pd.DataFrame(
    [
        {
            "name": name,
            "path": str(path),
            "exists": path.exists(),
        }
        for name, path in paths.items()
    ]
)

display(path_check)

missing = path_check[~path_check["exists"]]

if len(missing) > 0:
    raise FileNotFoundError("누락된 파일이 있습니다. 위 path_check를 확인하세요.")

phase_1a_metrics = pd.read_csv(paths["phase_1a_metrics"], encoding="utf-8-sig")
phase_1a_threshold = pd.read_csv(paths["phase_1a_threshold_metrics"], encoding="utf-8-sig")
phase_1a_predictions = pd.read_csv(paths["phase_1a_predictions"], encoding="utf-8-sig")

phase_2a_metrics = pd.read_csv(paths["phase_2a_metrics"], encoding="utf-8-sig")
phase_2a_threshold = pd.read_csv(paths["phase_2a_threshold_metrics"], encoding="utf-8-sig")
phase_2a_predictions = pd.read_csv(paths["phase_2a_predictions"], encoding="utf-8-sig")

same_vs_previous = pd.read_csv(paths["same_vs_previous_comparison"], encoding="utf-8-sig")
top_test_candidates = pd.read_csv(paths["top_test_candidates"], encoding="utf-8-sig")
bootstrap_summary = pd.read_csv(paths["bootstrap_summary"], encoding="utf-8-sig")

final_report_text = REPORT_PATH.read_text(encoding="utf-8")

shape_summary = pd.DataFrame(
    [
        {"name": "phase_1a_metrics", "shape": str(phase_1a_metrics.shape), "experiments": phase_1a_metrics["experiment_id"].nunique()},
        {"name": "phase_1a_threshold", "shape": str(phase_1a_threshold.shape), "experiments": phase_1a_threshold["experiment_id"].nunique()},
        {"name": "phase_1a_predictions", "shape": str(phase_1a_predictions.shape), "experiments": phase_1a_predictions["experiment_id"].nunique()},
        {"name": "phase_2a_metrics", "shape": str(phase_2a_metrics.shape), "experiments": phase_2a_metrics["experiment_id"].nunique()},
        {"name": "phase_2a_threshold", "shape": str(phase_2a_threshold.shape), "experiments": phase_2a_threshold["experiment_id"].nunique()},
        {"name": "phase_2a_predictions", "shape": str(phase_2a_predictions.shape), "experiments": phase_2a_predictions["experiment_id"].nunique()},
        {"name": "same_vs_previous", "shape": str(same_vs_previous.shape), "experiments": same_vs_previous["experiments"].sum() if "experiments" in same_vs_previous.columns else None},
        {"name": "top_test_candidates", "shape": str(top_test_candidates.shape), "experiments": top_test_candidates["experiment_id"].nunique()},
        {"name": "bootstrap_summary", "shape": str(bootstrap_summary.shape), "experiments": bootstrap_summary["experiment_id"].nunique()},
    ]
)

display(shape_summary)

print("final report chars:", len(final_report_text))
print("final report path:", REPORT_PATH)
```

```python
# 07 노트북 2번 셀: 최종 결론 요약표 생성
# 목적:
# - 최종 후보, sequence 비교 후보, previous-day 보수 후보를 한 표로 정리한다.
# - bootstrap CI까지 함께 붙인다.

selected_rows = [
    {
        "role": "best_same_date_classifier",
        "feature_timing": "same_date",
        "experiment_id": "phase1a_003",
        "interpretation": "현재 최고 성능의 same-date classifier",
    },
    {
        "role": "best_same_date_sequence_comparison",
        "feature_timing": "same_date",
        "experiment_id": "phase1a_002",
        "interpretation": "GRU 기반 sequence 비교 후보",
    },
    {
        "role": "same_date_window14_sensitivity",
        "feature_timing": "same_date",
        "experiment_id": "phase1a_007",
        "interpretation": "window 14에서 안정적인 MLP 비교 후보",
    },
    {
        "role": "best_previous_day_conservative_candidate",
        "feature_timing": "previous_day",
        "experiment_id": "phase2a_006",
        "interpretation": "strict previous-day timing 보수 후보",
    },
]

selected_df = pd.DataFrame(selected_rows)

threshold_all = pd.concat(
    [
        phase_1a_threshold.assign(feature_timing="same_date"),
        phase_2a_threshold.assign(feature_timing="previous_day"),
    ],
    ignore_index=True,
)

test_threshold_all = threshold_all[threshold_all["split"] == "test"].copy()

selected_metrics = selected_df.merge(
    test_threshold_all,
    on=["feature_timing", "experiment_id"],
    how="left",
)

ba_ci = bootstrap_summary[
    bootstrap_summary["metric"] == "balanced_accuracy"
][
    [
        "feature_timing",
        "experiment_id",
        "bootstrap_mean",
        "ci_lower_2_5",
        "ci_upper_97_5",
        "participants",
        "test_rows",
    ]
].rename(
    columns={
        "bootstrap_mean": "bootstrap_bal_acc_mean",
        "ci_lower_2_5": "bal_acc_ci_lower",
        "ci_upper_97_5": "bal_acc_ci_upper",
    }
)

roc_ci = bootstrap_summary[
    bootstrap_summary["metric"] == "roc_auc"
][
    [
        "feature_timing",
        "experiment_id",
        "bootstrap_mean",
        "ci_lower_2_5",
        "ci_upper_97_5",
    ]
].rename(
    columns={
        "bootstrap_mean": "bootstrap_roc_auc_mean",
        "ci_lower_2_5": "roc_auc_ci_lower",
        "ci_upper_97_5": "roc_auc_ci_upper",
    }
)

final_candidate_summary = (
    selected_metrics
    .merge(ba_ci, on=["feature_timing", "experiment_id"], how="left")
    .merge(roc_ci, on=["feature_timing", "experiment_id"], how="left")
)

final_candidate_summary = final_candidate_summary[
    [
        "role",
        "feature_timing",
        "experiment_id",
        "subset_name",
        "window",
        "model_family",
        "selected_threshold_from_validation",
        "balanced_accuracy",
        "bootstrap_bal_acc_mean",
        "bal_acc_ci_lower",
        "bal_acc_ci_upper",
        "roc_auc",
        "bootstrap_roc_auc_mean",
        "roc_auc_ci_lower",
        "roc_auc_ci_upper",
        "average_precision",
        "f1",
        "precision",
        "recall",
        "participants",
        "test_rows",
        "interpretation",
    ]
]

display(final_candidate_summary)

final_candidate_summary_path = (
    PHASE_2A_OUTPUT_ROOT / "final_candidate_summary_for_report.csv"
)

final_candidate_summary.to_csv(
    final_candidate_summary_path,
    index=False,
    encoding="utf-8-sig",
)

print("saved:", final_candidate_summary_path)
```

```python
# 07 노트북 3번 셀: 최종 한글 요약문 생성
# 목적:
# - 02_pipeline_from_scripts_summary.ipynb 또는 발표/보고서에 옮길 수 있는 최종 요약문을 만든다.
# - 모델 성능, 데이터 개선, previous-day 실험의 의미를 짧고 명확하게 정리한다.

best_row = final_candidate_summary[
    final_candidate_summary["role"] == "best_same_date_classifier"
].iloc[0]

sequence_row = final_candidate_summary[
    final_candidate_summary["role"] == "best_same_date_sequence_comparison"
].iloc[0]

previous_row = final_candidate_summary[
    final_candidate_summary["role"] == "best_previous_day_conservative_candidate"
].iloc[0]

summary_lines = []

summary_lines.append("# 최종 딥러닝 실험 요약")
summary_lines.append("")
summary_lines.append("## 1. 데이터 개선 요약")
summary_lines.append("")
summary_lines.append("- 기존 전체 feature 입력(`full_current`)은 validation 성능은 높았지만 test 일반화가 불안정했다.")
summary_lines.append("- sleep-session 계열, recovery 계열, same-date feature timing 위험을 점검했다.")
summary_lines.append("- 1차 개선에서는 feature subset ablation을 수행했다.")
summary_lines.append("- 2차 개선에서는 `target_date - 1 day` feature만 사용하는 previous-day dataset을 생성해 timing sensitivity를 확인했다.")
summary_lines.append("")
summary_lines.append("## 2. 최종 후보")
summary_lines.append("")
summary_lines.append(
    f"- 최종 1순위 모델은 `{best_row['feature_timing']} / {best_row['subset_name']} / "
    f"window {int(best_row['window'])} / {best_row['model_family']}`이다."
)
summary_lines.append(
    f"- test balanced accuracy는 `{best_row['balanced_accuracy']:.4f}`이고, "
    f"participant-level bootstrap 95% CI는 "
    f"`[{best_row['bal_acc_ci_lower']:.4f}, {best_row['bal_acc_ci_upper']:.4f}]`이다."
)
summary_lines.append(
    f"- test ROC AUC는 `{best_row['roc_auc']:.4f}`이고, "
    f"ROC AUC bootstrap 95% CI는 "
    f"`[{best_row['roc_auc_ci_lower']:.4f}, {best_row['roc_auc_ci_upper']:.4f}]`이다."
)
summary_lines.append("")
summary_lines.append("## 3. 딥러닝 sequence 비교 후보")
summary_lines.append("")
summary_lines.append(
    f"- sequence 모델 비교 후보는 `{sequence_row['feature_timing']} / {sequence_row['subset_name']} / "
    f"window {int(sequence_row['window'])} / {sequence_row['model_family']}`이다."
)
summary_lines.append(
    f"- test balanced accuracy는 `{sequence_row['balanced_accuracy']:.4f}`이며, "
    f"bootstrap 95% CI는 "
    f"`[{sequence_row['bal_acc_ci_lower']:.4f}, {sequence_row['bal_acc_ci_upper']:.4f}]`이다."
)
summary_lines.append("")
summary_lines.append("## 4. Previous-day 보수 실험")
summary_lines.append("")
summary_lines.append(
    f"- strict previous-day 후보는 `{previous_row['feature_timing']} / {previous_row['subset_name']} / "
    f"window {int(previous_row['window'])} / {previous_row['model_family']}`이다."
)
summary_lines.append(
    f"- test balanced accuracy는 `{previous_row['balanced_accuracy']:.4f}`이고, "
    f"bootstrap 95% CI는 "
    f"`[{previous_row['bal_acc_ci_lower']:.4f}, {previous_row['bal_acc_ci_upper']:.4f}]`이다."
)
summary_lines.append("- previous-day 실험에서는 성능이 크게 낮아졌으므로, 현재 최고 모델은 strict prior-day forecasting 모델로 해석하면 안 된다.")
summary_lines.append("")
summary_lines.append("## 5. 최종 해석")
summary_lines.append("")
summary_lines.append("- 현재 가장 강한 결과는 same-date daytime-only feature를 사용한 sleep classification이다.")
summary_lines.append("- same-date 성능은 높지만, sleep target의 `calendar_date`가 sleep end date 기준이므로 일부 feature timing 해석에 주의가 필요하다.")
summary_lines.append("- previous-day 실험은 더 보수적인 timing control이며, 이 조건에서는 예측력이 낮아졌다.")
summary_lines.append("- 따라서 최종 보고에서는 `same-date classification model`과 `previous-day timing sensitivity analysis`를 분리해서 제시하는 것이 적절하다.")
summary_lines.append("")

final_korean_summary = "\n".join(summary_lines)

print(final_korean_summary)

summary_path = PHASE_2A_OUTPUT_ROOT / "final_korean_experiment_summary.md"
summary_path.write_text(final_korean_summary, encoding="utf-8")

print("saved:", summary_path)
```
